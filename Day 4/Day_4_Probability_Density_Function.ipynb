**Probability density function (PDF) is used to define the random variable's probability coming within a distinct range of values**

PDF is density function which is calculated using Kernal Density Estimation technique. It tells in which range the density is more/less

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('iris.csv')
data
```

```python
data.rename(columns = {'SepalLengthCm':'SL', 'SepalWidthCm':'SW', 'PetalLengthCm':'PL', 'PetalWidthCm':'PW'}, inplace=True)
data.head()
```

```python
iris_setosa = data[data['Species'] == 'Iris-setosa']
iris_versicolor = data[data['Species'] == 'Iris-versicolor']
iris_virginica = data[data['Species'] == 'Iris-virginica']
```

```python
sns.FacetGrid(data, hue = 'Species', height = 6).map(sns.distplot, 'PL').add_legend()
```

```python
sns.FacetGrid(data, hue = 'Species', height = 6).map(sns.distplot, 'PW').add_legend()
```

```python
sns.FacetGrid(data, hue = 'Species', height = 6).map(sns.distplot, 'SL').add_legend()
```

```python
sns.FacetGrid(data, hue = 'Species', height = 6).map(sns.distplot, 'SW').add_legend()
```

We can see of all the 4 PDFs plotted, PL PDF is giving the best segregation bewteen the species. Hence we can use Petal Length PDF to segregate the species classes
